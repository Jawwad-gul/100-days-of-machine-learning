# Curse of Dimensionality  
Simply put, Having more and more Columns in data is bad news. (columns=dimensions)  
not putting simply, while creating more and more features (columns) does help the model to make more accurate predictions, but it can go so far. after a certain threshold (depends on model and data) increasing more features won't help in performace but it does increase the usage of computational resources as well as takes more time. One more thing is that it introduces the **Sparsity** in data. Sparsity in data refers to datasets where a significant number of values are zero or absent. when there are more features the chances of data being sparse are also more and hence it increases the distances between the actual values that are present, directly having a major affect on algorithms which work on distance based formulas i.e: KNN,linear-regression,logistic-regression. These all drawbacks are referred to as CoD (Curse of Dimensionality).<br>
