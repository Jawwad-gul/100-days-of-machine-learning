# **Transforming Categorical Data**
Data containing non-numeric stuff is called Categorical Data. There are two types of Categorical Data:
* **Nominal Data** (Categorical data which **is not** related to each other like: names, country, state, your engineering brach e.t.c)
* **Ordinal Data** (Categorical data which **is** related to each other... like classifying by a review : Excellent, Good, Bad: all are related)

### Encoding  
There are three ways to encode categorical data into numbers..:
1. **Ordinal Encoding**  (used for ordinal data, used for input columns)
2. **Label Encoding**  (almost similar to Ordinal Ecoding, its only used to encode the label/output column in classification problems)
3. **One-Hot Encoding**  (used for nominal data)
