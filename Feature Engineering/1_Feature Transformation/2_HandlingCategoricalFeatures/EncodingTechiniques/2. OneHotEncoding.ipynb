# **One Hot Encoding**
* Used for nominal categorical data.
* what it does: Makes a column for each category and when ever that category has to be encoded, it assigns 1 to that and 0 to others making a unique number. (>>> example in image)


![ytss](assets/OneHotEncoding.png)



Even if I have 50 different categories all occuring approximately equal number of times, OHE would still make 50 differnet columns for encoding them.



### Dummy Variable Trap  
ion know why its named like that... but basically what we do in it is we remove one column from input columns... lets say we remove Color_Y from above example, but if we remove, how would we differentiate between columns or even so how would we recognize any of them?.. Well taking a closer look we can see that each entry has only one "1" in them... like the code for red:[0,0,1], code for yellow is: [1,0,0] and for blue is: [0,1,0]. Now if we remove the Color_Y column these codes will look like yellow: [0,0], red:[0,1], blue: [1,0]. How can we differentiate: only column which has both 0s is yellow and other two are already unique so problem solved. we are able to differentiate even without the first column...  


Doing this also solved Multicollinearity, Multicollinearity occurs when input columns have some kind of mathematical relation between them. its a ML wisdom that the input should always be independent and label column should be dependent. if they are dependent it causes multicollinearity.  
In our example,  
before doing the dummy variable shii. was there some kind of mathematical relation coming up from input columns? yeah it was... adding them always makes "1", shown in Target column. so doing what we did just now also reduces Multicollinearity because some of the cells now will only have 0s in them making the sum/Target 0.  



### OHE Using Most Frequent Variables  
if we have a dataset (shape:(400,6)) of cars and there is a column named "brand" having 40 categories from which 5 categories are being repeated most of them time like 346 records are from these categories. then we encode these frequent categories with OHE and all of then other 35 categories are coupled into one let's say "others" reducing a lot of computational time for machine... because now we only have to create a 6 columns whereas earlier we had to create 40 columns.

## 4. OneHotEncoding for Top Categories

```python
counts = df['brand'].value_counts() #let's make a threshold for upto 100. meaning: every brand that has been recorded less then 100 times will be classified as other.
```

```python
type(counts)
```

```python
df['brand'].nunique()
threshold=100
```

```python
replace = counts[counts <= threshold].index
```

```python
pd.get_dummies(df['brand'].replace(replace,"other"))
```

```python
pd.get_dummies(df['brand'].replace(replace,"other")).replace({True:1,False:0})
```
