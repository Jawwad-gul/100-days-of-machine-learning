### Why?  
To get the data in Normal distributed form. What is **Normal Distribution**.  
_"The values are spread around the mean in a bell-shaped way"_   
bell-shaped mean when we plot the kde plot (`sns.kdeplot()`) the resultant pdf (probability density function) that gets plotted would be a bell-shaped shape... line going higher in the center and lower on rear sides.  
### Again Why?  
**Why do we need the data Normally Distributed?**  
Because the Linear models like Linear Rgression and Logistic Rgression expect the data to be Normally Distributed and perform well (give higher accuracy) when given the data that they expect.
