<br><br>  
There are two ways to handle misssing the values  
1. **Remove them** entirely. (just)
2. **Impute them.**

## Imputation  
There are two kinds of possible Imputation:
### 1. Univariate Imputation  
Imputing one column at a time. `SimpleImputer` class from `sklearn.impute` is used.  
There are two further types in univariate imputation based on data type.
   * ####  Numeric
      There are three different techniques to do numeric univariate imputation.
      1. **mean/media** (replace the missing values with mean or median)
      2. **Random** (replace with Random value from that column)
      3. **End of Distribution**
   * #### Categorical
     There are two techniques to handle missing values in categorical column
     1. **Mode** (replacing with most frequent value)
     2. **"Missing"** (replacing all the missing with any string literla like missing)
### 2. Multivariate Imputation  
Imputing multiple columns at a time.  
There are two techniques  
1. KNN Imputer (ML algorithm, `KNNImputer` class is used)
2. Iterative Imputer (`IterativeImputer` class is used)
