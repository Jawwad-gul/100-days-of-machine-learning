almost same as replacing the missing data with arbitrary number... the only difference is we use the formulas as described in the photo above `mean+3(std)` or `mean-3(std)` doing this takes gives us the last of the values of that column, aka outliers.
and then agian we can `df.fillna()` or make a `SimpleImputer(strategy='constant',fill_value=#one of the above)`  
this method of std and mean is used when data is normally distributed... if data is not that way.. then IQR approximity is used in it, there are also two values that we can replace with... first `Q1-1.5IQR` OR `Q3+1.5IQR` where `IQR = Q3-Q1` and Q1=25th percentile and Q3= 75th percentile.
