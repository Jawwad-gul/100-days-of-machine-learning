# Mixed Data  
When a column gas both numeric and categorical data  
Ther can be two cases: 
* "B5",this type of values where individual records have both type of data
* "A" and in next row but same columns "5", same column different data types of values

```python
import pandas as pd
import numpy as np
```

```python
df =  pd.read_csv('assets/titanic.csv')
df.info()
```

```python
df['number'].drop_duplicates()# how many people of family are traveling with this particular passenger.A means alone.
```

```python
fig = df['number'].value_counts().plot(kind='bar')
fig.set_title('Passengers traveling with')
```

```python
df['number_numerical'] = pd.to_numeric(df['number'],errors='coerce',downcast='integer')
```

```python
df['number_categorical'] = np.where(df['number_numerical'].isna(),df['number'],np.nan)
```

```python
df.head()
```

```python
df['Cabin'].unique()
```

```python
df['Ticket'].unique()
```

```python
df['cabin_num'] = df['Cabin'].str.extract('(\d+)')
df['cabin_cat'] = df['Cabin'].str[0]
```

```python
df.head()
```

```python
df['cabin_cat'].value_counts().plot(kind='bar')
```

```python
# last part of ticket as number
df['ticket_num'] = df['Ticket'].apply(lambda x: x.split()[-1])
df['ticket_num'] = pd.to_numeric(df['ticket_num'],errors='coerce',downcast='integer')

# first part of ticket as category
df['ticket_cat'] = df['Ticket'].apply(lambda x: x.split()[0])
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(),np.nan,df['ticket_cat'])

df.head()
```
