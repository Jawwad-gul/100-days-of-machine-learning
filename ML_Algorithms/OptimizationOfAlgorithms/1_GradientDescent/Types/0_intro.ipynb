#### Batch Gradient Descent  
When finding the slope to fill in the value of m and b we take in whole of the dataset for the calculation purposes. that can make the process slow, its rare , mostly only used when the function is convex. but also limitation is that data should't be much bigger.
#### Stochastic Gradient Descent  
It takes only one row to calculate the slope and then calculate the m and b making it faster but prone to errors.  
#### Mini Batch Gradient Descent  
Balanced between batch and stochastic GD. you decide a batch size. for example if the decided batch size is 30, then after every 30 rows visited in calculation, m and b will be updated. <br><br>

in B_DG, all of the rows are visited and then updated is made  
in S_GD, update is made after each row visited 
in M_GD, update is made after every batch-size rows.
