# Creating Manual Class for Gradient Descent

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_regression as mr
```

```python
class ManualGD:
    def __init__(self,b,m,lr,epochs):
        self.b = b
        self.m = m
        self.epochs = epochs
        self.lr = lr
    def fit(self,X,y):
        for i in range(self.epochs):
            loss_slop = -2*np.sum(y - self.m*X.ravel() - self.b)
            step_size = loss_slop*self.lr
            self.b = self.b - step_size
            y_pred = self.m*X + self.b
            plt.plot(X,y_pred)
        plt.scatter(X,y)
    def predict(self,X_test):
        pass
```

```python
X,y = mr( n_samples=4,n_features=1,n_targets=1,noise=20,random_state=13,n_informative=1)
```

```python
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X,y)
reg.coef_,reg.intercept_
```
