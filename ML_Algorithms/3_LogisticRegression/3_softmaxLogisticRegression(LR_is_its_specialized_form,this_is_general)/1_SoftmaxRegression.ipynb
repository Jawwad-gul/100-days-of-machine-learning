## Softmax Function  
Just like earlier we had sigmoid function. now we have softmax function. 

![ytss](assets/softmaxF.png)  
<br>
**Introduction**  <br>
for every class, its softmax is calculated through the given formula... its always between the range: 0 < softmax_value < 1. and the sum of softmax value of all the possible classes is always 1. if softmax function is calculated for just 2 classes its just as same as sigmoid function.

## Prediction  
<br>

![ytss](assets/softmaxP.png)

<br>

Now that we had three separate models each assigned to one out of three classes.  
1. **Model 1** will predict if the student is going to get the placement (1). otherwise (0) its not going to get the placement. its not concerned with what happens if its not getting placement.
2. **Model 2** will predict if the student will not get placement, others are just 0 for it.
3. **Model 3** will redict if the student is opting out (1) or not (0).  

**Process**:  
Now let's say that a new student comes in, data is given to each model. each model got the coefficients. so dot product the coefficients with the X (values of student.) to get the z, now its different for each model because each have different coefficients. now the value of z is given to softmax function separately for all three models to get the probability for all three classes separately. model 1's probability is for yes, model 2's proabability is for No and model 3's for optout. whichever one is larger becomes the final output  
<br>

## The Problem 
The problem here is that if data is larger applying one separate Logistic regression for each class becomes resource intensive as well slower. So how to make it so that it works with only one model.  

## The Solution:

## Code:  

```python
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.linear_model import LogisticRegression
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
encoder = LabelEncoder()
df['species'] = encoder.fit_transform(df['species'])
df.head()
```

```python
df = df[['sepal_length','petal_length','species']]
df.head()
```

```python
X = df.iloc[:,0:2]
y = df.iloc[:,-1]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
clf = LogisticRegression(multi_class='multinomial') # multinomial is other name of softmax regression
```

```python
clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
pd.DataFrame(confusion_matrix(y_test,y_pred))
```

```python
#prediction
query = np.array([[3.4,2.7]])
clf.predict_proba(query)
```
