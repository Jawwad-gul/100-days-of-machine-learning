## Perceptron Trick  
Gives the solution but not the best possible one.  
the equation here in logistic regression to find out the line is:  `Ax + By + c = 0` having 2d daa ofcourse.
<br>
![ytss](assets/perceptron.png)  
<br>

Suppose a dataset of students, we got cgpa and iq. Now there are students who got the placement and there are students who didn't get the placement. Our mission is to draw a line which can calssify the students based on placement. <br>
in perceptron trick, first random values are taken of coefficients, then using those values a line is drawn, and then by verifying through random data points the line is moved and it keeps happening until it reaches the desired location where it has all or atleast maximum number of students classified correctly (got placement on one side (say right), didn't get placement on left).<br> **How?** <br>
line always expects to have those students who got placement to be on its right side. so it will check from different data points and then if a certain data point is of the student who didn't get placement and its on the right side of the line... line will transform (change value of those coefficients) to the place where that point is on its left side, just like this it will keep transforming until it reaches the optmial position.

## How does transformation of line works?  
we do know this that line is formed by the coefficients. `A + Bx + C=0` here `A,B,C` are the coefficients and are the factors used to form the line. changing these will result in the transformed line. 
##### Changing `C`:  
Changing causes the line to move parallely on the y-axis. observe what happens when increased vs decreased: 
![ytss](assets/changingC.png)  
why increasing C causes the line to move downward? 
From  
Ax + By + C = 0,  
solve for y:  
y = -A/B * x - C/B.  

To find the y-intercept, set x = 0, so the slope term  
-A/B * x  becomes  -A/B * 0 = 0,  
leaving  
b = -C/B.  

When C increases, C/B increases, and the negative sign makes b more negative, so the line shifts downward. <br>

--- 

##### Changin `A`: 

![ytss](assets/changingA.png)  



For `Ax + By + C = 0`, solving for y gives:
`y = -(A/B)x - (C/B)`

When only A changes, the y-intercept `−C/B` stays fixed, so the line pivots around the same point.

The slope is `m = −A/B`, so increasing A makes the slope more negative, causing the line to tilt more sharply to the right while keeping the same y-intercept.

The x-intercept comes from setting `y = 0`:
`Ax + C = 0 → x = −C/A`

Since `C > 0` and `A > 0`, the result is always negative, so the x-intercept appears on the negative side.

Increasing A makes `−C/A` closer to 0, so the x-intercept moves rightward along the negative x-axis.

## Now we know how to move the line using A,B and C. Now, How to take it to the optimal postion?  
<br>

![ytss](assets/transform.png)  
<br>
1. identify the data points which are in wrong region w.r.t the current (intial: randomly selected) line.
2. if a data point is supposed to be in positive region and is in negative region, the tranformation has to occur which should result in that point being in the positive region of that transformed line.
3. suppose the point is `(4,5)`, add 1 within the coordinates --> `(4,5,1)`.
4. Subtract or add:
   * if the point is in positive region of current line. subtract these coordinates from the current line'e coefficients . `(A,B,C) - (4,5,1)`, these resultant numbers act as new coefficients.
   * if the point is in negative region of current line. add these coordinates to the current line'e coefficients . `(A,B,C) + (4,5,1)`, these resultant numbers act as new coefficients.
5. Now in this line the point `(4,5)` lies in the positive region of the line.

Now there is a problem with this approach... this approach causes huge transformation within one step. that's against machine learning practices. so first we multiply the coordinates with a learning-rate (which is usually 0.01 but can be changed), so that change is small

## Algorithm  
To get to the algorithm, there are certain changes in notations that need to be understood,
general equation: `Ax + By + C = 0`
becomes : `W0X0 + W1X1 + W2X2 = 0 ` where W0 = C and X0 is a column of all 1s in the beginning of data added manually. (why because it can be written in summition form) 
`SUMMITION(WiXi)` if it comes to be greater than 0 it means that model thinks that this data point should be on the positive side of the line. otherwise on the negative. 

![ytss](assets/algo.png)

<br>

**First check** : <br>
if the certain student belongs to negative region **and** model thinks it should belong to positive region   
Transform using: `Wnew = Wold - lr*Xi` <br>
**Second check** : <br>
if the certain student belongs to positive region **and** model thinks it should belong to negative region.  
Transform using: `Wnew = Wold + lr*Xi`

Transformations are only to be made when model-prediction (y_hat) and original data (y) differ: one is saying student got the placement while other opposes. that is why we previously added checks but now the same will be done without checks... because notice everytime the change is not required, in the formula `Wnew = Wold - 0` which keeps the old coefficient as they were.  
When the student belongs to those who didn't get the placement --> Negative, but lies on the positive side of line, Transformation must occur with - in between. and thats exactly how its happening in this case.. `y - y_hat` is handling it like a pro.
