# Code


```python
from sklearn.datasets import load_diabetes
import numpy as np
```

```python
data  = load_diabetes()
X = data.data
y = data.target
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=4)
```

```python
from sklearn.linear_model import Ridge
rig = Ridge(alpha=0.1,solver='cholesky') # this solver solves it by OLS, what we want
rig.fit(X_train,y_train)
print(rig.intercept_)
print(rig.coef_)
```

```python
from sklearn.metrics import r2_score
y_pred = rig.predict(X_test)
r2_score(y_test,y_pred)
```

```python
class ManualRidge:
    def __init__ (self,alpha=0.1):
        self.alpha = alpha
        self.coef_ = None
        self.intercept_ = None
    def fit(self,X_train,y_train):
        X_train = np.insert(X_train,0,1,axis=1)
        I = np.identity(X_train.shape[1])
        I[0][0] = 0 # because first column in X_Train are 1s those
        result = np.linalg.inv(np.dot(X_train.T,X_train) + self.alpha * I).dot(X_train.T).dot(y_train)
        self.intercept_ = result[0]
        self.coef_ = result[1:]
        pass
    def predict(self,X_test):
        return np.dot(X_test,self.coef_) + self.intercept_
```

```python
rg = ManualRidge()
rg.fit(X_train,y_train)
y_pred = rg.predict(X_test)
print(r2_score(y_test,y_pred))
print(rg.coef_,rg.intercept_)
```
