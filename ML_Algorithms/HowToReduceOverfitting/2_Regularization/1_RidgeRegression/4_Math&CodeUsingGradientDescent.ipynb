```python
from sklearn.datasets import load_diabetes
import numpy as np
```

```python
data = load_diabetes()
X = data.data
y = data.target
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=4)
```

```python
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import r2_score
sgd = SGDRegressor(alpha=0.001,eta0=0.1,learning_rate='constant',max_iter=500)

sgd.fit(X_train,y_train)
print('intercept',sgd.intercept_)
print('coef',sgd.coef_)
print('r2_score',r2_score(y_test,sgd.predict(X_test)))
```

```python
from sklearn.linear_model import Ridge
ri = Ridge(alpha=0.001,max_iter=500,solver='sparse_cg')
ri.fit(X_train,y_train)
print('intercept',ri.intercept_)
print('coef',ri.coef_)
print('r2_score',r2_score(y_test,ri.predict(X_test)))
```

```python
class ManualGD_Ridge:
    def __init__ (self,epochs,lr,alpha):
        self.epochs = epochs
        self.lr = lr
        self.alpha = alpha
        self.coef_ = None
        self.intercept_ = None
        

    def fit(self, X, y):
        # initialize
        self.coef_ = np.zeros(X.shape[1])
        self.intercept_ = 0

        for _ in range(self.epochs):

            # predictions
            y_pred = np.dot(X, self.coef_) + self.intercept_

            # error
            e = y_pred - y

            # gradient wrt w  (Xᵀ * X * w - Xᵀ * y + λw)
            dw = np.dot(X.T, np.dot(X, self.coef_)) - np.dot(X.T, y) + self.alpha * self.coef_

            # gradient wrt b  (only from error, no reg on b)
            db = np.sum(e)

            # update
            self.coef_ -= self.lr * dw
            self.intercept_ -= self.lr * db

            
    def predict(self,X_test):
        return np.dot(X_test, self.coef_) + self.intercept_
```

```python
    mr = ManualGD_Ridge(5000,0.001,0.01)
    mr.fit(X_train,y_train)
```

```python
y_pred = mr.predict(X_test)
r2_score(y_test,y_pred)
```
