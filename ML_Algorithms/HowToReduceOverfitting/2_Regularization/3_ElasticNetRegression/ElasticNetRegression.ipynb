when you want all of your features remaining intact in your data, you use ridge regression, when you know that there are some features which are less important and you want to remove them, thats when you use lasso... **but** when you don't know about your features that much, you can use Elastic Net regression.  
Its a combination of both ridge and lasso regression.
![ytss](assets/en.png)  
<br>
here in elasticnet regression there are two hyperparameters that can be controlled... lambda and l1-ratio, these are used to decide how much of both rigde and lasso regression are going to be used... 

```python
lam = a + b
l1_ratio =  a / lam
a = l1_ration * a # -------[1]
b = lam - a # -------[2]
```


ElasticNet proves to be better when the data has lot of multicollinearity (input cols are dependent on each other)


````python
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
````

````python
data = load_diabetes()
X = data.data
y = data.target
````

````python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=10)

````

````python
lr = LinearRegression()
lr.fit(X_train,y_train)
r2_score(y_test,lr.predict(X_test))
````

````python
la = Lasso(alpha=0.00001)
la.fit(X_train,y_train)
r2_score(y_test,la.predict(X_test))
````

````python
ri = Ridge(alpha = 0.00001)
ri.fit(X_train,y_train)
r2_score(y_test,ri.predict(X_test))
# very slight difference from lasso... depends on dataset
````

````python
el = ElasticNet(alpha=0.000001,l1_ratio=0.9)
el.fit(X_train,y_train)
r2_score(y_test,el.predict(X_test))
````
