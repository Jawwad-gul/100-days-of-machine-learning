```python
from sklearn.datasets import make_regression
import plotly.express as px
import plotly.graph_objects as go

import pandas as pd
import numpy as np

from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score 
```

```python
X,y = make_regression(n_samples=100,n_features=2,n_informative=2,n_targets=1,noise=50,random_state=42)
```

```python
df = pd.DataFrame({'feature_1':X[:,0],'feature_2':X[:,1],'target':y})
df.head()
```

```python
df.shape
```

```python
fig= px.scatter_3d(df,x='feature_1',y='feature_2',z='target')
fig.update_layout(
    height=)
fig.show()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=3)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
print("MAE: ",mean_absolute_error(y_test,y_pred))
print('MSA: ',mean_squared_error(y_test,y_pred))
print('R2 Square',r2_score(y_test,y_pred))
```
