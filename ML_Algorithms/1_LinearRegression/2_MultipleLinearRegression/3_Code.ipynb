# Manual Code vs Library code accuracy

### Library model accuracy:

```python
from sklearn.datasets import load_diabetes
import numpy as np
```

```python
data = load_diabetes()
```

```python
X = data.data
y = data.target
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=2,test_size=0.2)
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)

y_pred = lr.predict(X_test)
r2_score(y_test,y_pred)
```

```python
class ManualMLR:
    def __inti__(self):
        self.coef = None
        self.intercept = None
    def fit(self,X_train,y_train):
        X_train = np.insert(X_train,0,1,axis=1) # add 1 in the beginnning.. why? go to previous notebook


        # coef:
        betas  = np.linalg.inv(np.dot(X_train.T,X_train)).dot(X_train.T).dot(y_train)
        self.intercept = betas[0]
        self.coef = betas[1:]
        print(betas)
    def predict(self,X_test):
        y_pred = np.dot(X_test,self.coef) + self.intercept# y = B0 + B1X1 + B2X2...BnXn can be written in the dot product form... have done this in mathnotebook
        return y_pred
        
```

```python
mr = ManualMLR()
mr.fit(X_train,y_train)
```

```python
y_pred = mr.predict(X_test)
```

```python
r2_score(y_test,y_pred)#same
```
