# Why not to use Linear Regression?  
### Draw back of Multi Linear Regression.  
computaional complexity of calculating inverse of `(nxn)` matrix is **n^3**, which is huge. and in OLS ( Ordinary Least Square ), means we use formula to calcuate stuff... and in it we have to calculate inverse... so we hit the large computational cycle.  
the other method was Gradient Descent, it doesn't use mathematical formula way like its used in OLS
