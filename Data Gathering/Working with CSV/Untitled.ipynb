## opening csv from a URL

```python
import requests
from io import StringIO
```

```python
url = "https://raw.githubusercontent.com/google/dspl/master/samples/google/canonical/countries.csv"
header = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/141.0.0.0"}
req = requests.get(url,headers=header)
data = StringIO(req.text)
df = pd.read_csv(data)
df
```
